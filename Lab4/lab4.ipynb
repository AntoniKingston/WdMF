```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

# 1. A

```python
#na 1 ćwiartce bo nie robi to różnicy
def estimate_pi_primitive(n_samples):
    in_vec = [int((random.random()**2 + random.random()**2) < 1) for _ in range(n_samples)]
    cum_in = np.cumsum(in_vec)
    divisor = np.array([i+1 for i in range(n_samples)])
    pi_estimates = (cum_in / divisor) * 4
    return pi_estimates
```

```python
pic_moze = estimate_pi_primitive(10000)
plt.plot(pic_moze)
plt.axhline(y=np.pi, color='r', label='pi')
plt.xlabel('Number of samples')
plt.ylabel('Estimated value of pi')
```

# 1. B

```python
def estimate_pi_little_less_primitive(n_samples):
    rec_width = 1+1/np.sqrt(2)
    rec_height = 1-1/np.sqrt(2)
    in_vec = [int(((-1 + rec_width*random.random())**2 + (1/np.sqrt(2) + rec_height*random.random())**2) < 1) for _ in range(n_samples)]
    cum_in = np.cumsum(in_vec)
    divisor = np.array([i+1 for i in range(n_samples)])
    pi_estimates = 2 + 2 * (cum_in / divisor)
    return pi_estimates
```

```python
pic_moze = estimate_pi_little_less_primitive(10000)
plt.plot(pic_moze)
plt.axhline(y=np.pi, color='r', label='pi')
plt.xlabel('Number of samples')
plt.ylabel('Estimated value of pi')
```
