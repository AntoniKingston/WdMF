```python
import math
#x_0 is seed
def basic_linear_generator(n, x_0):
    a_1=math.floor(math.pi*10**9)
    a_0=math.floor(math.e*10**9)
    m = 2**35
    randoms = [x_0/(m-1)]
    curr = x_0
    for i in range(n):
        curr = (a_1*curr + a_0) % m
        randoms.append(curr/(m-1))
    return randoms
```

```python
randoms = basic_linear_generator(1000000, 37456374)
```

```python
import random
PI = math.pi
def f_byag(x):
    return (1+math.cos(2*PI*x))/2
def random_from_some_weird_dist(n):
    randoms = []
    while(len(randoms) < n):
        curr = []
        while(True):
            x = random.gauss(0,1)
            y = random.uniform(0,1)
            if y > f_byag(x):
                break
            curr.append(x)
        randoms += curr
    return randoms[:n]
    
```

```python
randoms = random_from_some_weird_dist(1000000)
```

### c1 z definicji konstrukcyjnej procesu

```python
def rand_pois(n, l=1):
    randoms = []
    for i in range(n):
        it = 0
        cumtime = 0
        while(True):
            cumtime += random.expovariate(l)
            if cumtime >= 1:
                randoms.append(it)
                break
            it+=1
    return randoms
```

```python
import numpy as np
l = 10
randoms = rand_pois(100000, l)
counts = np.bincount(randoms)
```

### c2 z własności infinitezymalnych

```python
def rand_pois2(n, l=1, dt=10**(-3)):
    randoms = []
    for i in range(n):
        jumps = 0
        for j in np.arange(0, 1, dt):
            x = random.uniform(0,1)
            if x < l*dt:
                jumps+=1
        randoms.append(jumps)
    return randoms
```

```python
l = 3
randoms = rand_pois2(100000, l)
counts = np.bincount(randoms)
```

```python
NORMALISING_CONST = 6/math.pi**2
def rand_from_another_weird_dist(n):
    randoms = []
    for i in range(1, n+1):
        #niby niezależność
        random.seed(i)
        x = random.uniform(0,1)
        cumsum = NORMALISING_CONST
        it = 1
        while(cumsum < x):
            it+=1
            cumsum += NORMALISING_CONST/it**2
        random.seed(int((1/i)*10**9) % n)
        #losowanie znaku
        x = random.uniform(0,1)
        if x < 0.5:
            it=-it
        randoms.append(it)
    return randoms
```

```python
from collections import Counter
randoms = rand_from_another_weird_dist(100000)
counts = Counter(randoms)
```

```python
def inverse_rising_approx(function, x):
    p = 0 
    l = 0
    while(function(p) < x):
        p+=10000
    while(function(l) > x):
        l-=10000
    while(p-l>10**(-5)):
        curr = (p+l)/2
        #na pewno się wydarzy xd
        if function(curr) == x:
            return curr
        elif function(curr) <= x:
            l=curr
        else:
            p=curr
    return curr
```

```python
from scipy.stats import norm

def bin101o3_dist(x):
    it = 0
    dist = 0
    p = 1/3
    jmp = 1-p
    if x >= 10:
        return 1
    while(it<=x):
        dist += math.comb(10,it)*p**it*jmp**(10-it)
        it+=1
    return dist
def distribuant(x):
    return 1/3*(bin101o3_dist(x) + norm.cdf(x,0,1) + ((1 - np.exp(-x)) if x>=0 else 0 ))
```

```python
def random_from_yet_another_weird_dist(n):
    randoms = []
    for i in range(n):
        x = random.uniform(0,1)
        randoms.append(inverse_rising_approx(distribuant,x))
    return randoms
```
